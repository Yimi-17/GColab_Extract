# Extraer Fecha, Titulo y URL

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

# Define la fecha de inicio y fin
fecha_inicio = datetime(2022, 3, 1)  # Fecha de inicio
fecha_fin = datetime(2023, 9, 21)  # Fecha de fin (actual)

# URL base
base_url = "https://pachamamaradio.org/{}/{:02d}/{:02d}/"

# Función para generar las URLs de las fechas
def generar_url(fecha):
    return base_url.format(fecha.year, fecha.month, fecha.day)

# Realiza un bucle para recorrer las fechas y raspar los datos
current_date = fecha_inicio
while current_date <= fecha_fin:
    url = generar_url(current_date)

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        target_div = soup.find("div", class_="td_block_inner tdb-block-inner td-fix-index")

        if target_div:
            h3_elements = target_div.find_all("h3", class_="entry-title td-module-title")

            if h3_elements:
                print(f"Fecha: {current_date.strftime('%Y-%m-%d')}")
                for h3_element in h3_elements:
                    a_element = h3_element.find("a")
                    if a_element:
                        titulo = h3_element.text.strip()
                        url_noticia = a_element.get("href")
                        print(f"Título: {titulo}")
                        print(f"URL de la noticia: {url_noticia}")

                        # Visita la URL de la noticia y extrae el primer párrafo
                        response_noticia = requests.get(url_noticia)
                        if response_noticia.status_code == 200:
                            soup_noticia = BeautifulSoup(response_noticia.text, "html.parser")
                            first_paragraph = soup_noticia.find("div", class_="tdb-block-inner td-fix-index").p
                            if first_paragraph:
                                print(f"Primer párrafo de la noticia: {first_paragraph.text.strip()}")
                            else:
                                print("No se encontró el primer párrafo en la noticia")
                        else:
                            print(f"No se pudo acceder a la URL de la noticia: {url_noticia}")
                    else:
                        print(f"No se encontró el enlace en la fecha {current_date.strftime('%Y-%m-%d')}")
            else:
                print(f"No se encontraron títulos en la fecha {current_date.strftime('%Y-%m-%d')}")
        else:
            print(f"No se encontró el div en la fecha {current_date.strftime('%Y-%m-%d')}")
    else:
        print(f"No se pudo acceder a la URL {url}. Estado de respuesta: {response.status_code}")

    # Avanza a la siguiente fecha
    current_date += timedelta(days=1)
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import re


# Define la fecha de inicio y fin año-mes-dia
fecha_inicio = datetime(2022, 3, 1)  # Fecha de inicio
fecha_fin = datetime(2023, 9, 21)  # Fecha de fin (actual)


# URL base
base_url = "https://pachamamaradio.org/{}/{:02d}/{:02d}/"



# Función para generar las URLs de las fechas
def generar_url(fecha):
    return base_url.format(fecha.year, fecha.month, fecha.day)


# Función para extraer el primer párrafo de una URL
def extraer_primer_parrafo(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        target_div = soup.find("div", class_="td_block_wrap tdb_single_content tdi_151 td-pb-border-top td_block_template_1 td-post-content tagdiv-type")

        if target_div:
            inner_div = target_div.find("div", class_="tdb-block-inner td-fix-index")

            if inner_div:
                first_p = inner_div.find("p")

                if first_p:
                    return first_p.text.strip()
                else:
                    return "No se encontraron elementos <p> dentro del div"
            else:
                return "No se encontró el div interno con la clase especificada"
        else:
            return "No se encontró el div class tdb-block-inner td-fix-index"
    else:
        return f"No se pudo acceder a la URL {url}. Estado de respuesta: {response.status_code}"


# Función para extraer la URL de la imagen de fondo
def extraer_url_imagen_fondo(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # Encuentra el div con la clase especificada
        # target_div = soup.find("div", class_="td_block_wrap tdb_single_bg_featured_image tdi_149 tdb-content-horiz-center td-pb-border-top td_block_template_1")
        target_div = soup.find("div", class_="td_block_wrap tdb_single_bg_featured_image tdi_148 tdb-content-horiz-center td-pb-border-top td_block_template_1")

        if target_div:
            # Busca las etiquetas <style> dentro del div
            style_tags = target_div.find_all("style")

            if style_tags:
                # Combinar el contenido de todas las etiquetas <style>
                combined_css_content = '\n'.join([style_tag.string for style_tag in style_tags])

                # Busca una URL en el contenido CSS combinado
                url_match = re.search(r'url\((.*?)\)', combined_css_content)

                if url_match:
                    found_url = url_match.group(1)
                    found_url = found_url.strip("'")  # Elimina comillas simples
                    print(f"IMG-URL: {found_url}")
                    # print("================================")
                else:
                    print("-- No se encontraron URLs en el CSS combinado")
            else:
                print("-- No se encontraron etiquetas <style> dentro del div")
        else:
            print("-- No se encontró el div class td_block_template_1")
    else:
        print(f"-- No se pudo acceder a la URL {url}. Estado de respuesta: {response.status_code}")



# Realiza un bucle para recorrer las fechas y raspar los datos
current_date = fecha_inicio
while current_date <= fecha_fin:
    url = generar_url(current_date)

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        target_div = soup.find("div", class_="td_block_inner tdb-block-inner td-fix-index")

        if target_div:
            h3_elements = target_div.find_all("h3", class_="entry-title td-module-title")

            if h3_elements:
                print(f"FECHA: {current_date.strftime('%Y-%m-%d')}")
                for h3_element in h3_elements:
                    a_element = h3_element.find("a")
                    if a_element:
                        titulo = h3_element.text.strip()
                        url_noticia = a_element.get("href")
                        print("============================================")
                        print(f"DATE: {current_date.strftime('%Y-%m-%d')}")
                        print(f"TITLE: {titulo}")
                        print(f"URL: {url_noticia}")


                        # Buscar el elemento <a> con la clase "td-post-category" un nivel arriba
                        categoria_element = a_element.find_previous("a", class_="td-post-category")
                        if categoria_element:
                            categoria = categoria_element.text.strip()
                            print(f"TAG: {categoria}")

                        # Extrae el primer párrafo utilizando la función
                        primer_parrafo = extraer_primer_parrafo(url_noticia)
                        print(f"ENTRY: {primer_parrafo}")

                        # Extrae la URL de la imagen de fondo utilizando la función
                        extraer_url_imagen_fondo(url_noticia)
                    else:
                        print(f"No se encontró el enlace en la fecha {current_date.strftime('%Y-%m-%d')}")
            else:
                print(f"No se encontraron títulos en la fecha {current_date.strftime('%Y-%m-%d')}")
        else:
            print(f"No se encontró el div en la fecha {current_date.strftime('%Y-%m-%d')}")
    else:
        print(f"No se pudo acceder a la URL {url}. Estado de respuesta: {response.status_code}")

    # Avanza a la siguiente fecha
    current_date += timedelta(days=1)
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import re

# Define la fecha de inicio y fin año-mes-dia
fecha_inicio = datetime(2022, 3, 1)  # Fecha de inicio
fecha_fin = datetime(2022, 3, 1)  # Fecha de fin (actual)

# URL base
base_url = "https://pachamamaradio.org/{}/{:02d}/{:02d}/page/{}/"

# Función para generar las URLs de las fechas y páginas
def generar_url(fecha, page):
    return base_url.format(fecha.year, fecha.month, fecha.day, page)

# Función para extraer el primer párrafo de una URL
def extraer_primer_parrafo(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        target_div = soup.find("div", class_="td_block_wrap tdb_single_content tdi_151 td-pb-border-top td_block_template_1 td-post-content tagdiv-type")

        if target_div:
            inner_div = target_div.find("div", class_="tdb-block-inner td-fix-index")

            if inner_div:
                first_p = inner_div.find("p")

                if first_p:
                    return first_p.text.strip()
                else:
                    return "No se encontraron elementos <p> dentro del div"
            else:
                return "No se encontró el div interno con la clase especificada"
        else:
            return "No se encontró el div class tdb-block-inner td-fix-index"
    else:
        return f"No se pudo acceder a la URL {url}. Estado de respuesta: {response.status_code}"

# Función para extraer la URL de la imagen de fondo
def extraer_url_imagen_fondo(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # Encuentra el div con la clase especificada
        target_div = soup.find("div", class_="td_block_wrap tdb_single_bg_featured_image tdi_148 tdb-content-horiz-center td-pb-border-top td_block_template_1")

        if target_div:
            # Busca las etiquetas <style> dentro del div
            style_tags = target_div.find_all("style")

            if style_tags:
                # Combinar el contenido de todas las etiquetas <style>
                combined_css_content = '\n'.join([style_tag.string for style_tag in style_tags])

                # Busca una URL en el contenido CSS combinado
                url_match = re.search(r'url\((.*?)\)', combined_css_content)

                if url_match:
                    found_url = url_match.group(1)
                    found_url = found_url.strip("'")  # Elimina comillas simples
                    print(f"IMG-URL: {found_url}")
                else:
                    print("-- No se encontraron URLs en el CSS combinado")
            else:
                print("-- No se encontraron etiquetas <style> dentro del div")
        else:
            print("-- No se encontró el div class td_block_template_1")
    else:
        print(f"-- No se pudo acceder a la URL {url}. Estado de respuesta: {response.status_code}")

# Realiza un bucle para recorrer las fechas y páginas y raspar los datos
current_date = fecha_inicio
while current_date <= fecha_fin:
    page = 1
    while True:
        url = generar_url(current_date, page)
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            target_div = soup.find("div", class_="td_block_inner tdb-block-inner td-fix-index")

            if target_div:
                h3_elements = target_div.find_all("h3", class_="entry-title td-module-title")

                if h3_elements:
                    print("///////////////////////////")
                    print(f"/// FECHA: {current_date.strftime('%Y-%m-%d')} ///")
                    print("///////////////////////////")
                    for h3_element in h3_elements:
                        a_element = h3_element.find("a")
                        if a_element:
                            titulo = h3_element.text.strip()
                            url_noticia = a_element.get("href")
                            print("============================================")
                            print(f"DATE: {current_date.strftime('%Y-%m-%d')}")
                            print(f"TITLE: {titulo}")
                            print(f"URL: {url_noticia}")

                            # Buscar el elemento <a> con la clase "td-post-category" un nivel arriba
                            categoria_element = a_element.find_previous("a", class_="td-post-category")
                            if categoria_element:
                                categoria = categoria_element.text.strip()
                                print(f"TAG: {categoria}")

                            # Extrae el primer párrafo utilizando la función
                            primer_parrafo = extraer_primer_parrafo(url_noticia)
                            print(f"ENTRY: {primer_parrafo}")

                            # Extrae la URL de la imagen de fondo utilizando la función
                            extraer_url_imagen_fondo(url_noticia)
                        else:
                            print(f"No se encontró el enlace en la fecha {current_date.strftime('%Y-%m-%d')}")
                else:
                    print(f"No se encontraron títulos en la fecha {current_date.strftime('%Y-%m-%d')}")
            else:
                print(f"No se encontró el div en la fecha {current_date.strftime('%Y-%m-%d')} - Página {page}")
                break  # No se encontró el div, sal del bucle de páginas
        else:
            print(f"##################################################")
            print(f"####### Bloque {url}. Respuesta: {response.status_code}")
            print(f"##################################################")
            break  # No se pudo acceder a la página, sal del bucle de páginas

        page += 1  # Avanza a la siguiente página

    # Avanza a la siguiente fecha
    current_date += timedelta(days=1)
```
